```python
noun=list() #initializing list for storing the nouns
determiner=['a','A','an','An','the','The', 'his','her','their']
for i in range(0,len(blake_poems_words)):
    if blake_poems_words[i] in determiner:
        noun.append(blake_poems_words[i+1])
```

```python
noun #list nouns
```
